```
import numpy as np
import pandas as pd

```

```
quotes_frame = pd.read_csv('quotes_inspirational.csv')
print(quotes_frame.shape)
print(quotes_frame.iloc[:3])
```

```
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```
text = nltk.word_tokenize("My name is Kyle and I like coding")
quote = nltk.word_tokenize(quotes_frame.iloc[0]['Quote'])
index = 0
for word in quote:
  if (word == "'ve"):
    quote[index] = "have"
  if (word == "'s"):
    quote[index] = "is"
  if (word == "'ll"):
    quote[index] = "will"
  if (word == "'t"):
    quote[index] = "not"
  if (word == "'re"):
    quote[index] = "are"
  if (word == "'d"):
    quote[index] = "had"
  
  index += 1

nltk.pos_tag(text)
nltk.pos_tag(quote)
```

```
from nltk.corpus import wordnet 
synonyms = [] 
  
for syn in wordnet.synsets("s"): 
    for l in syn.lemmas(): 
        synonyms .append(l.name()) 

print(synonyms) 
```

```
def tweak_quote(quote_index=0,num_changes=3):
  temp_quote = quotes_frame.iloc[quote_index]['Quote']
  temp_quote_token = nltk.word_tokenize(temp_quote)

  print(temp_quote)
  print(temp_quote_token)
  
  for i in range(num_changes):
    rand1 = np.random.randint(0,len(temp_quote_token))
    syns = []
    if temp_quote_token[rand1] in [',','.',':',';','\'']: continue
    for syn in wordnet.synsets(temp_quote_token[rand1]):
      for l in syn.lemmas():
        syns.append(l.name())
    if len(syns) is 0 : continue
    rand2 = np.random.randint(0,len(syns))
    temp_quote_token[rand1] = syns[rand2]

  print(" ".join(temp_quote_token))


tweak_quote(0,10)
  
```
